```python
from langchain_community.document_loaders import PyPDFLoader, csv_loader, json_loader, TextLoader, Docx2txtLoader
from pathlib import Path
```

```python
PARSERS = {
    ".pdf": PyPDFLoader,
    ".txt": TextLoader,
    ".csv": csv_loader,
    ".json": json_loader,
    ".docx": Docx2txtLoader,
}
```

```python
def parse_file(file_path: str, rows_per_block: int = 1) -> list[dict]:
    ext = Path(file_path).suffix.lower()
    parser = PARSERS.get(ext)
    if parser is None:
        raise ValueError(f"No parser registered for extension: {ext}")

    if ext in ROW_BATCHED_TYPES:
        return parser(file_path, rows_per_block)
    return parser(file_path)
```
